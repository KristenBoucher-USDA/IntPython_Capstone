### Objective:
The objective of my capstone project is to extract information from the Tradewinds online IT acquisition database (the Tradewinds Solutions Marketplace) and organize it so that IT personnel throughout the USDA can easily search for IT solutions that fit their needs. The end product is a searchable online database that houses the information from over 800 sources.

### Background

The Tradewinds Solutions Marketplace is a database that houses post-competition pitch videos for the digital space. It features products from a variety of vendors for AI, data analytics, GIS systems, and more. Since the items are post-competition, DOD and other entities can purchase the solutions directly without going through the market research and competitive process for procurement. This cuts down the amount of time for IT procurement significantly. 
The Marketplace is viewable once a government user registers and logs in. It is free to use and available to just about anyone. The challenge is that, since there are so many videos, it is very time consuming watching all of the videos to figure out whether or not a solution is right for a given problem. I was tasked with extracting all of the information from the Marketplace, including information from the video recordings, and putting it in one spreadsheet. I decided to take things one step further and create a web-based dashboard where people can search whatever terms they need and find information faster than if they logged in and searched the Marketplace themselves. Since the video transcriptions are already created, this eliminates the need to watch all of the five minute videos. 

The Tradewinds Solutions Marketplace can be foud at: https://www.tradewindai.com/tw-marketplace

### Explanation of Value:
Procurement of IT software and programming for the USDA historically is a very time consuming process. My team has been trying to find ways to expedite the process while also working within the confines of federal regulations and contracting requirements. Having access to the Tradewinds platform helps speed up the process by cutting out a large chunk of the market research and solicitation, thus streamlining procurement. However, the videos and vendor descriptions are kept behind a login and there are over 800 pages an videos for different products. My dashboard pulls all of the information from the web pages and videos and puts it into an accessible format for all of the mission areas and staff offices across the organization to search and locate IT software and systems that suit their needs.
